```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
graphs = []
max_degs = []
max_eigvals = []
for i in range(10):
    G = nx.generators.random_graphs.fast_gnp_random_graph(10, 0.5)
    graphs.append(G)
    graph_degrees = nx.degree(G)
    
    max_d = max(graph_degrees, key= lambda item: item[1])[1]
    max_degs.append(max_d)
    print(f"max degree: {max_d}")
    
    adj = nx.linalg.adjacency_matrix(G)
    adj = adj.todense()
    max_eigval = np.linalg.eigvals(adj).max()
    max_eigvals.append(max_eigval)
    print(f"max eig value: {max_eigval}\n")
    
plt.plot(max_eigvals,'g*', max_degs, 'ro')
plt.show()
```

```python
import random

def distribution_graph(g):
    print(nx.degree(g))
    all_node_degree = list(dict((nx.degree(g))).values())
  
    unique_degree = list(set(all_node_degree))
    unique_degree.sort()
    nodes_with_degree = []
    for i in unique_degree:
        nodes_with_degree.append(all_node_degree.count(i))
  
    plt.plot(unique_degree, nodes_with_degree)
    plt.xlabel("Degrees")
    plt.ylabel("No. of nodes")
    plt.title("Degree distribution")
    plt.show()

print("Enter number of nodes")
N = int(input())

print("Enter value of probability of every node")
P = float(input())

g1 = nx.Graph()

g1.add_nodes_from(range(1, N + 1))

for i in g1.nodes():
    for j in g1.nodes():
        if (i < j):

            R = random.random()

            if (R < P):
                g1.add_edge(i, j)
    pos = nx.circular_layout(g1)

nx.draw(g1, pos, with_labels=1)
plt.show()
  
# Display connection between nodes    
distribution_graph(g1)
```

```python
def p_i(g, ii):
    return g.degree[ii]/sum([g.degree[j] for j in range(g.number_of_nodes())])

def decision(probability):
    return random.random() < probability
```

```python
g = nx.generators.random_graphs.fast_gnp_random_graph(10, 0.4)
for i in range(10, 1000):
    g.add_node(i)
    for n in range(g.number_of_nodes()):
        if decision(p_i(g, n)):
            g.add_edge(i, n)

nx.draw(g,with_labels=1)
plt.show()
```
